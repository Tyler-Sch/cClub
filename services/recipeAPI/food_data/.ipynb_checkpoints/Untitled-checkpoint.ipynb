```python
import pandas as pd 
import numpy as np
from fuzzywuzzy import fuzz
```

```python
food = pd.read_csv('FOOD_DESC.csv')
```

```python
food.head(40)
```

```python
food.Long_Desc
```

```python
from fuzzywuzzy import process
```

```python
fuzz.token_sort_ratio('camembert cheese', 'Cheese, Camembert')
```

```python
choices = food.Long_Desc.values
```

```python
process.extract("kosher salt", choices, scorer=fuzz.token_set_ratio )
```

```python
l = ["1/4 cup (60ml) extra-virgin olive oil",
    "4 ounces (115g) pork belly, cut into 1-inch pieces",
    "Kosher salt",
    "4 garlic cloves, unpeeled and smashed",
    "4 ounces (115g) Spanish dry-cured chorizo, preferably younger and more tender, sliced into rounds",
    "1 red bell pepper, stemmed, seeded, and cut into 3/4-inch pieces",
    "1 Anaheim pepper, stemmed, seeded, and cut into 3/4-inch pieces]"
    ]
```

```python
for i in l:
    print(i)
    print(process.extract(i, choices, scorer=fuzz.token_set_ratio))
```
